## Audio Features

**Acousticness** describes whether the song uses primarily acoustic instruments or electronic/electric instruments. A value of 1.0 indicates that a song is purely acoustic. <br> 

**Danceability** describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythmic stability, beat strength, and overall regularity, with a value of 1.0 indicating high danceability. <br> 

**Duration (ms)** simply indicates the length of the song in milliseconds.<br> 

**Energy** is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. <br> 

**Instrumentalness** predicts whether a track contains no vocals. “Ooh” and “aah” sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly “vocal”. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. <br> 	

**Liveness** detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. <br>

**Loudness (dB)** of a recording is measured in decibels (dB). Loudness values are averaged across the entire song and typically range between -60dB and 0dB.  <br>

**Mode** indicates the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.<br>

**Popularity** of a song is a metric of how often the song has been streamed on Spotify，but with more recent streams weighted more heavily. <br>

**Speechiness** detects the presence of spoken words in a recording. The more exclusively speech-like the recording (ex. talk shows, audio books, etc.), the closer the speechiness is to 1.0.

**Tempo** is the speed of the song, or the number of beats per minute (BPM). <br>

**Time signature** is a notational convention to specify how many beats are in each bar of music. <br>

**Valence** describes the musical positiveness conveyed by the song. In other words, songs with high valence sound more happy/positive/cheerful, while songs with low valance sound more negative/depressed/angry. 

```python
import json
import requests
from bs4 import BeautifulSoup
import string
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error 
from wordcloud import WordCloud
```

# Top Hits

```python
with open('../data/top_hits_merged_clean_lyrics_audio_features.json') as json_file:
    tracks = json.load(json_file)
    
df = pd.read_json(tracks)
```

```python
df.shape
```

```python
df['year']=df['album'].apply(lambda x: x.get('release_date')[:4])
df['decade'] = df['year'].apply(lambda x: x[:3]+'0')
df['primary_artist'] = df['artists'].apply(lambda x: x[0]['name'])
df['artists'] = df['artists'].apply(lambda x: [artist['name'] for artist in x])
```

```python
df.columns
```

### Distributions of Spotify Audio Features

Below you can see the distributions of the Audio Features from Spotify. 
<br>

**Acousticness** - Most of the Top Hit songs have a low acousticness, probably due to the fact of the emergence of electronic-based beats/instrumentals in the music industry during the last few decades. <br> 

**Danceability** - The danceability of Top Hit songs is normally distributed.  <br> 

**Duration (ms)** - Most songs are between 3-5 minutes long.  <br> 

**Energy** - Top Hit songs are skewed towards having higher energy. <br> 

**Instrumentalness** - Almost all Top Hit contain vocal content. <br> 	

**Liveness** - It appears that most Top Hit songs are studio-recorded. <br>

**Loudness (dB)** - The loudness of Top Hit songs is normally distributed. <br>

**Mode** - Almost 2/3 of the Top Hit songs are in a major key. <br>

**Popularity** - ?.<br>

**Speechiness** ?.

**Tempo** - It appears to be normally distributed around 120BPM. <br>

**Time signature** - Almost all Top Hit songs have 4 beats per bar, but a small number of hits have 3 beats per bar, probably slower ballads. <br>

**Valence** - Top Hit songs are more likely to have high valance. <br>


# Naturally, no Hot 100 singles have a speechiness of 1.0, but a significant number have non-zero speechiness, most likely rap songs.

# speechness is high in a minority of our songs, which means that the majority of our songs do not contain a lot of text/lyrics; this result is pulled up by the minority of rap/hip-hop songs that by essence contain a lot of text 


```python
ax = df[['acousticness', 'danceability', 'duration_ms', 'energy',
           'instrumentalness', 'liveness', 'loudness', 'mode',
           'popularity', 'speechiness', 'tempo', 'time_signature', 'valence']].\
            hist(figsize=(16,12), bins=20, alpha=0.75);
ax = ax.ravel()
plt.tight_layout()

titles = ['Acousticness', 'Danceability', 'Duration (miliseconds)', 'Energy',
          'Instrumentalness', 'Liveness', 'Loudness (dB)', 'Mode',
          'Popularity', 'Speechiness', 'Tempo', 'Time Signature', 'Valence']

for i in range(len(ax)-3):
    ax[i].set_title(titles[i])
    if i % 4 == 0:
        ax[i].set_ylabel('Number of Songs')
plt.suptitle('Distributions of Spotify Audio Features', fontsize=20, y=1.05);
plt.show()
```

### Correlation heatmap 

From the correlation heatmap we can see:

**Positive correlations:**
- The two strongest correlated features are energy and loudness. Typically, energetic tracks feel fast, loud, and noisy, so this positive correlation is expected. The correlation graph of these two features is shown down below.
- The second strongest correlated features are energy and acousticness.
- The third strongest corelated features are danceability and valence, which means that the happier is the song, the more danceble it is.

**Negative correlations:**
- The strongest negative correlation is between energy and acousticness, basicly lound songs aren't acoustic. The correlation graph of these two features is shown down below.
- The second strongest negative correlation is achieved between loudness and acousticness, which means that the louder the song, the less acoustic it is.
- The third strongest negative correlation is between valence and acousticness, wich means that the sad songs are often acoustic.


```python
num_df_attributes = df[['acousticness', 'danceability', 'duration_ms', 'energy',
           'instrumentalness', 'liveness', 'loudness', 'mode',
           'popularity', 'speechiness', 'tempo', 'time_signature', 'valence']]

num_df_attributes.corr('spearman')['popularity']
```

```python
fig, ax = plt.subplots(figsize=(12,12))   
sns.heatmap(
    num_df_attributes.corr(), 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True,
    ax = ax
)
```

```python
ax = sns.regplot(x=df['energy'], y=df['loudness'], line_kws= {'linewidth': 1, 'color': 'red'}, scatter_kws={'s': 1})
ax.set(title='Correlation: 0.68')
ax.set(xlabel='Energy', ylabel='Loudness (dB)')
plt.show()
ax1 = sns.regplot(x=df['energy'], y=df['acousticness'], line_kws= {'linewidth': 1, 'color': 'red'}, scatter_kws={'s': 1})
ax1.set(title='Correlation: -0.58')
ax1.set(xlabel='Energy', ylabel='Acousticness')
plt.show()
```

### Boxplots over decades

From these boxplots we can identify a few trends:
- acousticness seems to have decreased as well over time, probably due to the emergence of electronic-based musics which contain less real acoustic instruments that "traditional" songs
- duration has slightly decreased since the 70's: we can indeed state that more and more songs are short than before
- energy has increased over time
- speechness has increased over time, certainly due to the emergence of rap/hip-hop music, which contain a lot of text by essence

```python
audio_features=['acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness',
                'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
```

```python
for feature in audio_features:
    fig, ax = plt.subplots(figsize=(8,8))
    plt.suptitle('')
    ax.grid(False)
    df.boxplot(column=[feature], by='decade', ax=ax, showfliers=False, grid=False)
    plt.show()
```

### Most populart artists

It looks like Michael Jackson and Madonna are still the artists which have the most hit songs for the past 40 years, followed by Elton John and Mariah Carey. 
The fact that only "old" artists are the most popular can be explained by the rising concurrence between artists: as there are more artists today than before, it was easier to stay in the top list back in the days than today (more songs can challenge the top hits today than before).

### Most ________________ artist

In the word clouds below we can see the leading artists who have highest values for each audio feature. 

```python
# when creating word clouds, maps word fontsize to a color
class colormap_size_func(object):
    def __init__(self, colormap, max_font_size):
        import matplotlib.pyplot as plt
        self.colormap = plt.cm.get_cmap(colormap)
        self.max_font_size = max_font_size

    def __call__(self, word, font_size, position, orientation,
                 random_state=None, **kwargs):
        if random_state is None:
            random_state = Random()
        r, g, b, _ = 255 * np.array(self.colormap(font_size / self.max_font_size))
        return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)
```

```python
top_artists = df.groupby('primary_artist')['name'].count()
top_artists = top_artists[top_artists >= 5]
audio_features=['acousticness', 'danceability', 'energy','instrumentalness', 
                'liveness', 'loudness', 'speechiness', 'tempo', 'valence']

X_artists = df.groupby('primary_artist')[audio_features].mean()
X_artists = X_artists.reindex(top_artists.index)
```

```python
# make subplot axes
fig, axes = plt.subplots(3, 3, figsize=(16,15))
axes = axes.ravel()
adjective = ['acousticness', 'danceability', 'energy','instrumentalness', 
            'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
colormaps = ['Purples', 'OrRd', 'cool', 'spring_r', 'Greens', 'winter_r', 'GnBu',
             'PuRd', 'autumn_r']

for i in range(len(audio_features)):
    # make word-weight dictionary
    color_func = colormap_size_func(colormaps[i], 200)
    feature = X_artists[audio_features[i]] + 1e-3 # no zero values allowed
    dictionary = dict(zip(X_artists.index, feature))
    
    # make word cloud
    wordcloud = WordCloud(background_color='white', height=1200,
                          min_font_size=5, width=1400, color_func=color_func,
                          max_words=len(X_artists), relative_scaling=0.5,
                          max_font_size=300)
    wordcloud = wordcloud.generate_from_frequencies(dictionary)
    
    # display word cloud
    axes[i].imshow(wordcloud, interpolation='bilinear')
    axes[i].axis('off')
    axes[i].set(title='most {} artists'.format(adjective[i]).title())
```
