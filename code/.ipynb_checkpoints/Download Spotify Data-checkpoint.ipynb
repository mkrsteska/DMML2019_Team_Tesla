```python
def save_track_and_features(results, store, found):
    for i, item in enumerate(results['items']):
        try: 
            track = item['track']
            song_title = track['name']
            artist_name = track['artists'][0]['name']
            current_song = (song_title, artist_name)
            if current_song in found: continue
            audio_features = sp.audio_features(track['id'])[0]
            track.update(audio_features)
            store.append(track)
            found.add(current_song)
        except:
            continue
```

```python
def get_top_hits():
    i = 0
    res = []
    found = set()
    playlist = sp.user_playlist('spotify', playlist_id='3cS7RrbJgQPEiWPAyOnb4l')
    tracks = playlist['tracks']
    save_track_and_features(tracks, res, found)
    while tracks['next']:
        i+=1
        print(i)
        tracks = sp.next(tracks)
        save_track_and_features(tracks, res, found)
    return res
```

```python
hits = get_top_hits()
```

```python
def get_songs(playlists):
    res = []
    found = set()
    for i, playlist_item in enumerate(playlists):
        print(i)
        playlist = sp.user_playlist('spotify', playlist_id=playlist_item)
        tracks = playlist['tracks']
        save_track_and_features(tracks, res, found)
        while tracks['next']:
            tracks = sp.next(tracks)
            save_track_and_features(tracks, res, found)
    return res
```
